```python
acoes = ['BBAS3','TAEE4','BRAP4']

url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol=ITUB3.SAO&apikey={API_KEY}&datatype=csv'
r = requests.get(url) 
df = pd.read_csv(StringIO(r.text))
```

```python
compilado = pd.DataFrame()

for acao in acoes:
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={acao}.SAO&apikey={API_KEY}&datatype=csv'
    r = requests.get(url) 
    df = pd.read_csv(StringIO(r.text))
    compilado = pd.concat([compilado,df])
```

```python
compilado.head()
```

```python
msg = ''
for _,row in compilado.iterrows():
    msg += f'{row['symbol']} está custando {row['price']} \n'

print(msg)
```
